```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
d1 = '/home/u0890475/Documents/SW_analysis/20251110_amr0_xshort'
```

```python
def get_dydu(df):
    xvel_list=np.array(df['x_velocity'])
    t_list =df['time_s']
    z_list = df['z_m']
    dz = z_list[0]
    t = t_list[0]
    return np.mean(xvel_list)/dz,t
    
```

```python
file_list = []
for root,_,files in os.walk(d1):
    for file_name in files:
        file_list.append(file_name)


```

```python
file_list.sort()
sl_ext= {'p40':1.853043,'p45':2.270543}
thml_diff = {'p40':9.807093e-5,'p45':1.036447e-4}
file_list
num_f  = len(file_list)
dt40 =[]
dc40 =[]
dt45=[]
dc45=[]
t40 =[]
t45=[]
Da40=[]
Da45=[]
```

```python
for i in range(num_f):
    temp_df = pd.read_csv(d1+'/'+file_list[i])
    g,t = get_dydu(temp_df)
    print(file_list[i])
    if 'p40' in file_list[i]:
        dt40.append(math.sqrt(thml_diff['p40']/g))
        dc40.append(sl_ext['p40']/g)
        Da40.append(sl_ext['p40']**2 /(thml_diff['p40'] * g))
        t40.append(t)
    if 'p45' in file_list[i]:
        dt45.append(math.sqrt(thml_diff['p45']/g))
        dc45.append(sl_ext['p45']/g)    
        Da45.append(sl_ext['p45']**2 /(thml_diff['p45'] * g))
        t45.append(t)
```
